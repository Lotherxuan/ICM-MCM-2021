```python
import folium
import pandas as pd
cdata=pd.read_csv('../data/viirs-snpp_2019_Australia.csv')
cdata
```

```python
limit=200
data=cdata.iloc[0:limit,:]
lat=-37
lon=144
incidents=folium.map.FeatureGroup()
for latitude,longitude in zip(data.latitude,data.longitude):
    incidents.add_child(folium.CircleMarker([latitude,longitude], radius=7, color='red',fill=True,fill_color='red',fill_opacity=0.4))
m=folium.Map(location=[lat,lon])
m.add_child(incidents)
```
